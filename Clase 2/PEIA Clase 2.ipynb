```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

# Ejercicio 1

```python
n=1000
u =stats.uniform.rvs(size=n)
print(u[:10])
```

```python
x = np.ceil(u*6)
print(x[:10])
```

```python
plt.bar(range(1,7), [np.sum(x==i)/n for i in range(1,7)])
```

# Ejercicio 2

```python
t = -5*np.log(1-u)
```

```python
sns.histplot(t, stat='density')
tt = np.linspace(0,35, 1000)
plt.plot(tt, stats.expon.pdf(tt, scale=5),linewidth = 3, color='C1')
```

```python
plt.plot(tt, stats.expon.cdf(tt, scale=5),linewidth = 3, color='C1')
```
