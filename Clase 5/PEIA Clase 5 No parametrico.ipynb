# Estimación no paramétrica

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv('advertising.csv')
sales.head()
```

```python
newspaper = sales.Newspaper
max(newspaper)
```

```python
# plot ECDF
sns.ecdfplot(newspaper)

# add median (plug-in estimator)
median=newspaper.median()
plt.axhline(0.5,color='r', linestyle='--')
plt.axvline(x=median,color='r', linestyle='--')
plt.annotate(f"median={median}",(median+1,0.5-.05))
```

```python
# convergence of ECDF
for k in range(10,len(newspaper),10):
    sns.ecdfplot(newspaper[:k],label=k)
```

```python
sns.histplot(newspaper, stat='density')
```

```python
sns.kdeplot(newspaper)
```
