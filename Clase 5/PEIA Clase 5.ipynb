# Estimación no paramétrica

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv('advertising.csv')
sales.head()
```

```python
newspaper = sales.Newspaper
max(newspaper)
```

```python
# plot ECDF
sns.ecdfplot(newspaper)

# add median (plug-in estimator)
median=newspaper.median()
plt.axhline(0.5,color='r', linestyle='--')
plt.axvline(x=median,color='r', linestyle='--')
plt.annotate(f"median={median}",(median+1,0.5-.05))
```

```python
# convergence of ECDF
for k in range(10,len(newspaper),10):
    sns.ecdfplot(newspaper[:k],label=k)
```

```python
sns.histplot(newspaper, stat='density')
```

```python
sns.kdeplot(newspaper)
```

# Intervalos de confianza
## Ej 4:

```python
from scipy import stats
from numpy import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
q95=stats.norm.ppf(0.95)
print(q95)
```

```python
np.random.seed(2)
x = stats.norm.rvs(size=20, loc=3, scale=2)
x
```

```python
x.mean() - q95*2/sqrt(20)
```

## Obs:t de Student converge a la normal

```python
x = np.linspace(-4,4,1000)
```

```python
plt.plot(x, stats.norm.pdf(x), color='C0')
plt.plot(x, stats.t.pdf(x,df=1), color='C1')
plt.plot(x, stats.t.pdf(x,df=5), color='C2')
plt.plot(x, stats.t.pdf(x,df=10), color='C3')
plt.plot(x, stats.t.pdf(x,df=30), color='C4')
```

## Ej 5:

```python
t995= stats.t.ppf(0.999,df=49)
t995
```

```python
x = stats.norm.rvs(size=50, loc=2, scale=3)
sns.histplot(x)
x.mean()
```

```python
print(f"Intervalo de confianza: ({x.mean()-t995*x.std()/sqrt(50)},{x.mean()+t995*x.std()/sqrt(50)} )")
```

```python
print(f"Longitud del Intervalo: {2*t995*x.std()/sqrt(50)}")
```
