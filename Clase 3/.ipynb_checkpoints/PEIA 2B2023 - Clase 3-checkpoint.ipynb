# Ej 4: mejor estimador lineal (LMMSE) para X=U(0,1) e Y=X^2

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# genero X e Y
X=np.random.rand(100)
Y=X**2

# calculo var y cov
muX   =1/2
varX  =1/12
muY   =1/3
varY  =4/45
covXY =1/12

# defino estimador LMMSE de Y basado en X (rojo)
aX=covXY/varX
bX=muY-aX*muX
RX=aX*X+bX


# defino estimador LMMSE de X basado en Y (negro)
aY=covXY/varY
bY=muX-aY*muY
RY=aY*Y+bY
```

```python
print(f"corr(X,Y)={covXY/np.sqrt(varX)/np.sqrt(varY):.2f}")
print(f"g(X)={aX:.2f}X+{bX:.2f}")
print(f"g(Y)={aY:.2f}Y+{bY:.2f}")
```

```python
# estimador LMMSE de Y basado en X
plt.scatter(X,Y)
plt.plot(X[:10],RX[:10],'r-^')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
#estimador LMMSE de X basado en Y
plt.scatter(Y,X)
plt.plot(Y[:10],RY[:10],'k-^')
plt.xlabel('Y')
plt.ylabel('X')
```

## Regresión lineal para predecir Y a partir de X (continuación)

```python
XX=np.hstack((np.ones((100,1)),X.reshape(-1,1)))
YY=Y.reshape(-1,1)
hatBeta=np.linalg.inv(XX.T@XX)@XX.T@YY
```

```python
hatBeta
```

```python
hatY=XX@hatBeta
```

```python
plt.scatter(X,Y)
plt.plot(X[:10],hatY[:10],'g-^')
plt.xlabel('X')
plt.ylabel('Y')
```
